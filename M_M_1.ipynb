# M/M/1 Queueing model


```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
tt = int(input('Enter Total time in hours = '))
arr_rate = int(input('Enter arrival rate in (/hour) = '))
ser_rate = int(input('Enter service rate in (/hour) = '))

# Here Ideal condition is ser_rate > arr_rate always
```

```
n = int(np.random.poisson(arr_rate)*tt)   
```

```
n
```

```
inter_arr_time = []
arr_time = []
ser_time = []
```

##calculating inter arrival time

```
for i in range(n):
    temp = int(np.random.exponential(1/arr_rate)*60*60)
    if i==0:
        inter_arr_time.append(0)
    else:
        inter_arr_time.append(temp)
```

```
np.transpose(inter_arr_time)
```

##calculating arrival time

```
for i in range(n):
    if i==0:
        arr_time.append(0)
    else:
        arr_time.append(arr_time[i-1]+inter_arr_time[i])
```

```
np.transpose(arr_time)
```

##calculating service time

```
while not len(ser_time) == n:
    temp = int(np.random.exponential(1/ser_rate)*60*60)
    if not temp<1:
        ser_time.append(temp)
```

```
np.transpose(ser_time)
```

##Trafic Intensity


```
rho = arr_rate/ser_rate
```

```
rho
```

#limiting distribution

```
pi = []
for j in range(n):
    p = (1-rho)*(rho**j)
    pi.append(p)

# pi_j, pi^_j, pi*_j, p_j all are equal for M/M/1 and M/M/s
```

```
np.transpose(pi)
```

##The average number of customers in the system


```
Ls = rho/(1-rho)
```

```
Ls       
```

##The average number of customer spends in the queue

```
Lq = Ls - rho
```

```
Lq
```

##The average time a customer spends in the system

```
Ws = Ls/arr_rate
```

```
Ws
```

##The average time a customer spends in the queue

```
Wq = Ws - (1/ser_rate)
```

```
Wq
```

```
fig = plt.figure(figsize=(15,15))
ax = fig.subplots(2,2)
ax[0,0].plot([i for i in range(n)], pi, 'bo-' )
ax[0,0].set_title('Limiting distribution')
ax[0,1].plot([i for i in range(n)], inter_arr_time, 'go-')
ax[0,1].set_title('Inter arrival time')
ax[1,0].plot([i for i in range(n)], arr_time, 'or-')
ax[1,0].set_title('Arrival time')
ax[1,1].plot([i for i in range(n)], ser_time, 'yo-')
ax[1,1].set_title('Service time')
plt.show;
```
